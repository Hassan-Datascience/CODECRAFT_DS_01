```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('INTERNSHIP/API_SP.POP.TOTL_DS2_en_csv_v2_174326.csv', skiprows=4)
metadata = pd.read_csv('INTERNSHIP/Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_174326.csv')

countries_only = metadata[metadata['Region'].notna()]['Country Code']

df_filtered = df[df['Country Code'].isin(countries_only)]

top_10 = df_filtered.nlargest(10, '1960')

plt.figure(figsize=(12, 6))
plt.bar(range(len(top_10)), top_10['1960'], color='coral', edgecolor='black', alpha=0.7)

plt.xlabel('Country Name')
plt.ylabel('Population')
plt.title('Top 10 Actual Countries by Population (1960)')

plt.xticks(range(len(top_10)), top_10['Country Name'], rotation=45, ha='right')

plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```
